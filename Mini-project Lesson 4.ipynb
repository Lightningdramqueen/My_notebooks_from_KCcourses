```python
#теперь вытаскиваем полные пути к папкам выше
path = 'shared/homeworks/python_ds_miniprojects/4/data'
list1 = []
for i in range(len(os.listdir('shared/homeworks/python_ds_miniprojects/4/data'))):
    list1.append(path + '/' + os.listdir(path)[i])
list1
```

```python
#а теперь к папкам в этих папках, копаем глубже
list2 = []
for i in list1:
    for j in os.listdir(i):
        list2.append(i+"/"+j)
list2
```

```python
#а теперь уже к самим файлам в этих папках, мы у цели
list3 = []
for i in list2:
    for j in os.listdir(i):
        list3.append(i+"/"+j)
list3
```

```python
#и соединяем все в один датафрейм, цель достигнута
df = pd.DataFrame()
for i in range(len(list3)):
    df1 = pd.read_csv(list3[i])
    df1['name'] = list3[i].split('/')[6]
    df1['date'] = list3[i].split('/')[5]
    df = df.append(df1)
df    
```

```python
df.dtypes
  
```

```python
# переводим столбец даты в дату
df['date'] = pd.to_datetime(df.date)
```

```python
# убираем лишний столбец
df = df.drop(["Unnamed: 0"], axis=1)
```

```python
df.quantity.sum()
```

```python
#какой пользователь купил больше всего товаров
df.groupby('name', as_index=False) \
    .agg({'quantity':'sum'}) \
    .sort_values(['quantity'], ascending=False)
```

```python
#топ-10 товаров по числу проданных единиц за всё время
top_product = df.groupby('product_id', as_index=False) \
    .agg({'quantity':'sum'}) \
    .sort_values(['quantity'], ascending=False)
top_product
```

```python
#давайте визуализируем предыдущий шаг
import matplotlib as plt
%matplotlib inline 
plt.rcParams['figure.figsize']=20,10
import seaborn as sns




ax = sns.barplot(x="product_id", y="quantity", data=top_product, alpha = 0.5, color = 'green')
```

```python
#Посмотрим на продажи по дням! 
top_date = df.groupby('date', as_index=False) \
    .agg({'quantity':'sum'}) \
    .sort_values(['quantity'], ascending=False)
top_date
```

```python
#меняем формат даты, иначе на графике некрасиво
top_date['date'] = pd.to_datetime(top_date["date"].dt.strftime('%m/%d/%Y'))
top_date
```

```python
#все равно вышло некрасиво 
ax = sns.barplot(x="date", y="quantity", data=top_date, alpha = 0.7, color = 'blue' )
```
