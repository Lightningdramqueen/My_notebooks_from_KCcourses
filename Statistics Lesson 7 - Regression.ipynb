```python
#укажем верные названия машин
cars['CarNameNew'] = cars['CarName'].apply(lambda x: x.split(' ')[0])
```

```python
#и удалим лишний столбец
cars = cars.drop(['CarName'], axis = 1)
```

```python
cars['CarNameNew'] = cars['CarNameNew'].apply(lambda x: x.lower())
```

```python
#последний штрих в приглаживании этой колонки
cars['CarNameNew'] = cars['CarNameNew'].replace(['maxda', 'porcshce', 'toyouta', 'vokswagen', 'vw'], ['mazda', 'porsche', 'toyota', 'volkswagen', 'volkswagen'])
```

```python
colums = ['CarNameNew', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower']
```

```python
#отберем только те колонки, которые будут нам нужны для регрессии
cars_regr = cars[['CarNameNew', 'fueltype', 'aspiration','carbody', 'drivewheel', 'wheelbase', 'carlength','carwidth', 'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower','price']]
```

```python
#сделаем дамми-переменные чтобы включить категориальные переменные в нашу модель
f_dummy = pd.get_dummies(data=cars_regr[['CarNameNew', 'fueltype', 'aspiration','carbody', 'drivewheel', 'enginetype', 'cylindernumber']], drop_first = True)
```

```python
f_dummy
```

```python
#соединяем таблицу с категориальными переменными с таблицей с количественными переменными
final_for_regression = pd.concat([cars_regr[['wheelbase', 'carlength','carwidth', 'curbweight', 'enginesize', 'boreratio','horsepower','price']], f_dummy], axis=1)
```

```python
final_for_regression
```

```python
final_for_regression.shape
```

```python
import statsmodels.formula.api as smf 
#предскажем цену машины с помощью такого предиктора, как кол-во лошадиных сил - для примера
results = smf.ols('price ~ horsepower', final_for_regression).fit()
print(results.summary())
```

```python
#теперь соберем все предикторы в список, чтобы было удобнее включить их все в модель
final_for_regression.columns.tolist()
```
