```python
#переводим дату в верный формат
london['timestamp'] = pd.to_datetime(london['timestamp'], format='%Y-%m-%d %H:%M:%S')
```

```python
london = london.sort_values('timestamp')
```

```python
#делаем дату индексом чтобы применить resample по дням
london_resample = london.set_index('timestamp')
```

```python
london_resample
```

```python
#resample и строим график
london_resample.resample(rule='D').cnt.sum().plot()
```

```python
#сделаем из него датафрейм
new_resample = pd.DataFrame(london_resample.resample(rule='D').cnt.sum())
new_resample = new_resample.reset_index()
```

```python
new_resample
```

```python
#считаем скользящее среднее для числа аренд
new_resample['rolling_mean'] = new_resample.rolling(window=3).mean()
```

```python
new_resample
```

```python
#считаем разницу между средним и наблюдаемым
new_resample['difference'] = new_resample.cnt - new_resample.rolling_mean
```

```python
new_resample
```

```python
#считаем среднеквадр. отклонение
std = np.std(new_resample.difference)
```

```python
std
```

```python
#задаем доверительный интервал
new_resample['upper_bound'] = new_resample.rolling_mean + 2.576*std
```

```python
new_resample['lower_bound'] = new_resample.rolling_mean - 2.576*std
```

```python
new_resample
```

```python
#выделяем аномально высокие значения
new_resample.query('cnt > upper_bound')
```

```python
#выделяем аномально низкие значения
new_resample.query('cnt < lower_bound')
```
