```python
#приводим названия колонок к нужному формату
bookings.columns = [x.lower() for x in bookings.columns]
bookings.columns = bookings.columns.str.replace(" ", "_")
```

```python
bookings_head= bookings.head(7)
```

```python
#Пользователи из каких стран совершили наибольшее число успешных бронирований
successbook = bookings \
    .query("is_canceled == 0") \
    .groupby(['country'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) 
```

```python
successbook.head(5)
```

```python
#На сколько ночей в среднем бронируют отели
meannights = bookings \
    .groupby(['hotel'], as_index=False) \
    .aggregate({'stays_total_nights': 'mean'}) 
```

```python
meannights['stays_total_nights'].round(2)
```

```python
#На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017?
datestravel = bookings \
    .query("is_canceled == 0") \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
```

```python
datestravel
```

```python
#на какой месяц бронирования отеля типа City Hotel отменялись чаще всего в 2015? 2016? 2017? 
cancelled = bookings \
    .query("is_canceled == 1") \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .aggregate({'is_canceled': 'count'}) \
    .sort_values('is_canceled', ascending=False) \
```

```python
#сколько всего детей в отеле
bookings = bookings.assign(total_kids=bookings.loc[:, 'children':'babies'].sum(1))
```

```python
#а где в среднем больше?
children = bookings \
    .groupby(['hotel'], as_index=False) \
    .aggregate({'total_kids': 'mean'}) \
    .sort_values('total_kids', ascending=False) \
```

```python
children['total_kids'] = children['total_kids'].round(2)
```

```python
children
```

```python
bookings['has_kids'] = bookings.total_kids > 0 
```

```python
bookings
```

```python
#показатель оттока среди пользователей без детей

nokids_churn = bookings.query('is_canceled == 1 and has_kids == False').shape[0] / bookings.query('is_canceled == 0 and has_kids == False').shape[0]
nokids_churn = round(nokids_churn*100,2)
```

```python
nokids_churn
```

```python
#показатель оттока среди пользователей с детьми

yeskids_churn = bookings.query('is_canceled == 1 and has_kids == True').shape[0] / bookings.query('is_canceled == 0 and has_kids == True').shape[0]
yeskids_churn = round(yeskids_churn*100,2)
```
